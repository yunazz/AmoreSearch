## 회사 NEWS 카테고리 크롤링

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 기본 URL
base_url = "https://www.apgroup.com"
page_url_template = "https://www.apgroup.com/int/ko/news/news,1,list1,{}.html"

start_page = 2
end_page =145  # 마지막 페이지

# 데이터 저장 리스트
news_data = []

# 페이지 순회하면서 크롤링
for page_no in range(start_page, end_page + 1):
    url = page_url_template.format(page_no)
    print(f"크롤링 중: {url}")
    
    try:
        response = requests.get(url)
        response.encoding = 'utf-8' 
        response.raise_for_status()  
    except requests.exceptions.RequestException as e:
        print(f"페이지 {page_no}에서 오류 발생: {e}")
        continue

    soup = BeautifulSoup(response.text, "html.parser", )
    news_items = soup.select(".news-list li")  # 뉴스 리스트

    for item in news_items:
        post_ctgry = item.select_one(".thumb-desc .label.one")  # 카테고리
        title = item.select_one(".thumb-desc h3")  # 제목
        created_at = item.select_one(".thumb-desc .date")  # 날짜
      
        if post_ctgry and title and created_at:
            news_data.append({
                "post_ctgry": post_ctgry.get_text(strip=True),
                "title": title.get_text(strip=True),
                "created_at": created_at.get_text(strip=True),
            })

    time.sleep(1)  # 서버 부담을 줄이기 위해 1초 대기

# CSV로 저장
df = pd.DataFrame(news_data)
csv_filename = "ap_news.csv"
df.to_csv(csv_filename, index=False, encoding="utf-8")

print(f"크롤링 완료! CSV 파일 저장됨: {csv_filename}")
```
